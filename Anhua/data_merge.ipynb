```python
import numpy as np
import pandas as pd
```

```python
## load in acs data
df_acs = pd.read_csv("./acs_data/ACS_data.txt", sep = "\t")

## load in aqi data
df_aqi = pd.read_csv("./aqi_data/aqi_data_clnd.csv")
df_aqi["NAME"] = df_aqi.County + " County, " + df_aqi.State

## load in crime data
df_crime = pd.read_csv("./crime_data/crime_data_clnd.csv")
df_crime = df_crime.rename(index = str, columns = {"FIPS_ST": "state", "FIPS_CTY":"county"})

### Merging three datasets

## 1. merging aqi and acs
df_merged = pd.merge(df_acs, df_aqi, how = "left", on = ["NAME"])
## Since we have fewer counties in air quality data, we will use the state_average to replace it
## col_to_replace = df_merged.filter(like='Days ratio', axis=1).columns
## df_merged.loc[pd.isnull(df_merged['Unnamed: 0_y']),col_to_replace] = \

## merging acs and crime data
df_merged = pd.merge(df_merged, df_crime, how =  "left", on  = ["state", "county"])
    

```
