```python
import numpy as np
import pandas as pd
```

```python
temp_stations = pd.read_excel("temp-inventory.xlsm",index_col = 0)
column_names = ['latitude','longitude','elevation','state','name']
temp_stations = temp_stations[column_names]

prcp_stations = pd.read_excel("prcp-inventory.xlsm",index_col = 0)
column_names = ['latitude','longitude','elevation','state','name']
prcp_stations = prcp_stations[column_names]

snow_stations = pd.read_excel("snow-inventory.xlsm",index_col = 0)
column_names = ['latitude','longitude','elevation','state','name']
snow_stations = snow_stations[column_names]


```

```python
mly_tavg = pd.read_excel("mly-tavg-normal.xlsm",index_col = 0)
mly_tmax = pd.read_excel("mly-tmax-normal.xlsm",index_col = 0)
mly_tmin = pd.read_excel("mly-tmin-normal.xlsm",index_col = 0)
mly_htdd = pd.read_excel("mly-htdd-normal.xlsm",index_col = 0)
mly_cldd = pd.read_excel("mly-cldd-normal.xlsm",index_col = 0)
mly_prcp= pd.read_excel("mly-prcp-normal.xlsm",index_col = 0)
```

```python
mly_snow = pd.read_excel("mly-snow-normal.xlsm",index_col = 0)
```

```python
def get_monthly_temp(df):
    for index,row in df.iterrows():
        for i in range(12):
            lenth = len(row[i])-1
            data = row[i][:lenth]
            if data == '-7777':
                row[i] = '0'
            elif data in ['-9999','-6666','-5555']:
                row[i] = 'NA'
            else:
                temp = round(int(data) * 0.1,1)
                row[i] = str(temp)
    return df

def get_monthly_prcp(df):
    for index,row in df.iterrows():
        for i in range(12):
            lenth = len(row[i])-1
            data = row[i][:lenth]
            if data == '-7777':
                row[i] = '0'
            elif data in ['-9999','-6666','-5555']:
                row[i] = 'NA'
            else:
                prcp = round(int(data) * 0.01,2)
                row[i] = str(prcp)
    return df

def get_monthly_snow(df):
    for index,row in df.iterrows():
        for i in range(12):
            lenth = len(row[i])-1
            data = row[i][:lenth]
            if data == '-7777':
                row[i] = '0'
            elif data in ['-9999','-6666','-5555']:
                row[i] = 'NA'
            else:
                snow = round(int(data) * 0.1,1)
                row[i] = str(snow)
    return df

def get_monthly_dd(df):
    for index,row in df.iterrows():
        for i in range(12):
            lenth = len(row[i])-1
            data = row[i][:lenth]
            if data == '-7777':
                row[i] = '0'
            elif data in ['-9999','-6666','-5555']:
                row[i] = 'NA'
            else:
                dd = int(data)
                row[i] = str(dd)
    return df
```

```python
average_temp = get_monthly_temp(mly_tavg)
```

```python
average_temp
```

```python
max_temp = get_monthly_temp(mly_tmax)
min_temp = get_monthly_temp(mly_tmin)
month_prcp = get_monthly_prcp(mly_prcp)
month_cdd = get_monthly_dd(mly_cldd)
month_hdd = get_monthly_dd(mly_htdd)
```

```python
month_snow = get_monthly_snow(mly_snow)
```

```python
match_mth_prcp = month_prcp.loc[average_temp.index]
```

```python
match_mth_prcp
```

```python
index = match_mth_prcp.index
```

```python
columns = match_mth_prcp.columns
```

```python
climate_df = pd.DataFrame(index = index, columns = columns)
```

```python
index_snow = month_snow.index
for index, row in climate_df.iterrows():
    for i in range(12):
        temp_avag = average_temp.loc[index][i]
        temp_max = max_temp.loc[index][i]
        temp_min = min_temp.loc[index][i]
        cdd = month_cdd.loc[index][i]
        hdd = month_hdd.loc[index][i]
        prcp = match_mth_prcp.loc[index][i]
        if index in index_snow:
            snowfall = month_snow.loc[index][i]
        else:
            snowfall = '0.0'
        row[i] = (temp_avag,temp_max,temp_min,cdd,hdd,prcp,snowfall)
```
